```python
# Installation des dépendances
!pip install spacy pyspellchecker
!python -m spacy download fr_core_news_sm

# Importation des librairies nécessaires
import spacy
from google.colab import files
from spellchecker import SpellChecker

# Chargement du modèle français de spaCy
nlp = spacy.load("fr_core_news_lg")

# Téléchargement du fichier
uploaded = files.upload()

# Lecture du document texte
document_text = next(iter(uploaded.values())).decode('utf-8')

# Traitement du texte avec spaCy
doc = nlp(document_text)

# Extraction des mots et vérification orthographique
spell = SpellChecker(language='fr')
words = [token.text for token in doc if token.is_alpha]  # Extraire les mots
misspelled = spell.unknown(words)  # Identifier les mots mal orthographiés

# Calcul des statistiques
total_words = len(words)
errors_count = len(misspelled)
error_percentage = (errors_count / total_words) * 100

# Affichage des résultats
print(f"Nombre total de mots: {total_words}")
print(f"Nombre d'erreurs détectées: {errors_count}")
print(f"Pourcentage d'erreurs: {error_percentage:.2f}%")
print("Mots incorrectement identifiés:", misspelled)
```
